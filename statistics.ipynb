```python
measures = {
    "age": [15, 18, 25, 25, 40, 55, 58, 60, 80],
    "height": [160, 162, 165, 168, 172, 174, 174, 174, 176]
}
```

```python
# df_measures = pd.DataFrame(measures)
df_measures = pd.DataFrame.from_dict(measures)
```

```python
df_measures
```

## Resumos das medidas

### Medidas de posição

```python
# Média
df_measures['age'].mean()
```

```python
# Mediana
df_measures['age'].median()
```

```python
# Moda
df_measures['age'].mode()
```

### Medidas de dispersão

```python
# Variância
df_measures['age'].var()
```

```python
# Desvio padrão
df_measures['age'].std()
```

```python
# Coeficiente de variação - Idade
df_measures['age'].std() / df_measures['age'].mean() * 100
```

```python
# Coeficiente de variação - Altura
df_measures['height'].std() / df_measures['height'].mean() * 100
```

### Medidas de forma

#### Assimetria

A assimetria é uma medida de forma que indica a simetria da distribuição dos dados. Se a distribuição é simétrica, a assimetria é zero. Se a distribuição é assimétrica à direita, a assimetria é positiva. Se a distribuição é assimétrica à esquerda, a assimetria é negativa.

#### Curtose

A curtose é uma medida de forma que indica o pico da distribuição dos dados. A curtose é comparada com a curtose da distribuição normal, que é zero. Se a curtose é maior que zero, a distribuição é mais alta e mais afunilada que a distribuição normal. Se a curtose é menor que zero, a distribuição é mais baixa e mais achatada que a distribuição normal.

```python
# Assimetria - Idade
df_measures['age'].skew()
```

```python
# Simetria - Altura
df_measures['height'].skew()
```

```python
# Curtose - Idade
df_measures['age'].kurtosis()
```

```python
# Curtose - Altura
df_measures['height'].kurtosis()
```

```python
# Obter medidas estatísticas de todas as colunas
df_measures.describe()
```

### Sales DataFrame

## Desafio

Com base nas aulas do módulo de Estatística, utilize os métodos da biblioteca Pandas para tratar o dicionário abaixo e:

- Trazer a média das vendas
- Criar um gráfico de barras vertical mostrando o mês de referência e o valor
- Criar um gráfico de linhas mostrando o mês de referência e o valor.

```python
dict_faturamento = {
    "data_ref": [
        "2023-01-01",
        "2020-02-01",
        "2021-03-01",
        "2022-04-01",
        "2023-05-01",
        "2023-06-01",
        "2020-07-01",
        "2021-08-01",
        "2022-09-01",
        "2023-10-01",
        "2022-11-01",
        "2023-12-01",
    ],
    "valor": [
        400000,
        890000,
        760000,
        430000,
        920000,
        340000,
        800000,
        500000,
        200000,
        900000,
        570000,
        995000,
    ],
}

df_faturamento = pd.DataFrame(dict_faturamento)
df_faturamento["data_ref"] = pd.to_datetime(df_faturamento["data_ref"])
```

```python
df_faturamento
```

```python
# Trazer a média das vendas
df_faturamento["valor"].mean()
```

```python
# Criar um gráfico de barras vertical mostrando o mês de referência e o valor
df_faturamento.sort_values("data_ref").plot.bar(
    x="data_ref",
    y="valor",
    title="Faturamento por mês",
    xlabel="Mês de referência",
    ylabel="Faturamento",   
)
```

```python
# Criar um gráfico de linhas mostrando o mês de referência e o valor.
df_faturamento.sort_values("data_ref").plot.line(
    x="data_ref",
    y="valor",
    title="Faturamento por mês",
    xlabel="Mês de referência",
    ylabel="Faturamento",   
)
```
